# Data Formatting (categorical)


In this formatting tutorial we will see the categorical case.


Let's get some data:

```python
#Importacion de pandas
import pandas as pd

data = pd.read_excel('https://github.com/DiegoAntonio-web/PC2/blob/main/HDI%20(1).xlsx?raw=true')
data
```

```python
data_limpia = data.iloc[3:]
data_limpia
```

```python
# Eliminacion de columnas que tienen nombres NaN en su primera fila, excepto las 2 primeras
cols_to_keep = data_limpia.columns[:2] #Conservacion de las 2 primeras columnas
data_final = pd.concat([data_limpia[cols_to_keep], data_limpia.loc[:, data_limpia.iloc[0].notna()].iloc[:, 2:]], axis=1)

data_final
```

```python
#Eliminacion de todas las filas que contengan valores NaN a excepcion de la segunda columna
data_depurada = data_final[~(data_final.iloc[:, 2:].isna().all(axis=1) & data_final.iloc[:, 1].notna())]

data_depurada
```

```python
#Eliminacion de filas que contengan solo valores NaN

data_depurada = data_depurada.dropna(how='all')
data_depurada
```

```python
data_depurada = data_depurada.fillna('')

data_depurada
```

```python
data_depurada = data_depurada.replace({'..': '', '—': ''})
data_depurada
```
