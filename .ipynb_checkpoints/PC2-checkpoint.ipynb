```python
#Importacion de pandas
import pandas as pd
# Obtención de la tabla de Excel sin encabezado
data = pd.read_excel('https://github.com/DiegoAntonio-web/PC2/blob/main/HDI%20(1).xlsx?raw=true', header=None)

# Seleccionar las filas 5, 6 y 7 como nuevo encabezado (índices 4, 5 y 6)
new_header = data.iloc[4:7]  # Filas 5, 6 y 7

# Eliminar las filas usadas como encabezado
data = data.drop(index=[4, 5, 6])

# Combinar las filas seleccionadas en un nuevo encabezado
data.columns = new_header.apply(lambda x: ' '.join(x.dropna().astype(str)), axis=0)

# Restablecer el índice
data.reset_index(drop=True, inplace=True)

# Mostrar el DataFrame resultante
data
```

```python
# Eliminar las primeras 4 filas
data_1 = data.drop(index=[0, 1, 2, 3])

# Restablecer el índice
data.reset_index(drop=True, inplace=True)

# Mostrar el DataFrame resultante
data_1
```

```python
#Eliminacion de todas las filas que contengan valores NaN a excepcion de la segunda columna
data_2 = data_1[~(data_1.iloc[:, 2:].isna().all(axis=1) & data_1.iloc[:, 1].notna())]

data_2
```

```python
# Eliminación de columnas que contienen NaN en la primera fila
data_2 = data_2.loc[:, data_2.iloc[0].notna()]

# Reiniciar el índice del DataFrame
data_2.reset_index(drop=True, inplace=True)

# Mostrar el DataFrame resultante
data_2
```

```python
# Eliminación de filas donde el primer elemento de la primera columna es NaN
data_2 = data_2.dropna(subset=[data_2.columns[0]], axis=0)

# Reiniciar el índice del DataFrame
data_2.reset_index(drop=True, inplace=True)
data_2
```

```python
data_2.info()
```

```python
# Convertir las columnas de la 3 a la 7 a float64
for col in data_2.columns[2:7]:  # Índices de columna 2 a 6 (3 a 7)
    data_2[col] = pd.to_numeric(data_2[col], errors='coerce')  # Convertir a float64

# Verificar el resultado
data_2
```

```python
data_2.info()
```

```python
# Arreglamos el error de lectura de la columna 7, ya que el python lo leyo como "," 1,500 del excel, aca lo lee como 1500.0
# Dividir todos los elementos de la columna 7 entre 1000
data_2[data_2.columns[6]] = data_2[data_2.columns[6]] / 1000

# Verificar el resultado
data_2
```
